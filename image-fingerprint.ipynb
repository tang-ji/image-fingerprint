```python
from PIL import Image
import imagehash
import matplotlib.pyplot as plt
```

```python
image_file1 = "imgs/sunrise.jpg"
image_file2 = "imgs/sunrise2.jpg"
image_file3 = "imgs/baseball.png"
```

```python
def get_hash(image_file):
    image = Image.open(image_file)
    return str(imagehash.dhash(image))

def distance(hash1, hash2):
    c = 0
    for i in range(len(hash1)):
        if hash1[i] != hash2[i]:
            c += 1
    return c

def compare(image_file1, image_file2):
    f, ax = plt.subplots(1, 2, figsize=(10,5), dpi=80)
    hash1 = get_hash(image_file1)
    hash2 = get_hash(image_file2)
    ax[0].set_title(hash1)
    ax[0].imshow(plt.imread(image_file1))
    ax[0].axis('off')

    ax[1].set_title(hash2 + "(distance:{})".format(distance(hash1, hash2)))
    ax[1].imshow(plt.imread(image_file2))
    ax[1].axis('off')
```

```python
compare(image_file1, image_file2)
```

```python
compare(image_file1, image_file3)
```
